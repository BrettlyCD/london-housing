**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(b):
    """Taking in the borough (b), calculate the mean price average from 2018 and 1998 and calculate the ratio between the two"""
    y2018 = properties[(properties['borough'] == b) & (properties['year'] == 2018)].groupby('year')['price_avg'].mean().iloc[0]
    y1998 = properties[(properties['borough'] == b) & (properties['year'] == 1998)].groupby('year')['price_avg'].mean().iloc[0]
    ratio = y2018 / y1998
    return ratio
```

```python
ratio_dict = {} #create blank dictionary
```

```python
#iterate through array of boroughs and save the 2018 to 1998 ratio for each to the ratio_dict dictionary
for b in properties['borough'].unique():
    key = b
    ratio_dict[key] = create_price_ratio(b)
```

```python
#load the ratio_dict dictionary to a pandas dataframe
price_ratio = pd.DataFrame(ratio_dict.items(), columns=['london_borough', 'ratio_price'])

#set borough to index
price_ratio = price_ratio.set_index('london_borough')
```

```python
price_ratio.head()
```

```python
#sort from highest to lowest
price_ratio = price_ratio.sort_values('ratio_price', ascending=False)
```

```python
#select the 15 boroughs with the highest increases
price_ratio_15 = price_ratio.sort_values('ratio_price', ascending=False).head(15)
```

```python
price_ratio_15.plot(kind='bar')
plt.title('Top 15 London Boroughs- 2018 vs 1998 Mean Home Price Ratio')
plt.xlabel('Borough')
plt.ylabel('2018:1998 Mean Price Ratio')
plt.show()
```

```python
price_ratio.describe()
```

```python
total_graph = properties.groupby('year')['price_avg'].mean()
max_graph = properties[properties['borough'] == price_ratio.head(1).index.values[0]].groupby('year')['price_avg'].mean()
min_graph = properties[properties['borough'] == price_ratio.tail(1).index.values[0]].groupby('year')['price_avg'].mean()
```

```python
total_graph.plot(kind="line", label = 'All Boroughs')
max_graph.plot(kind="line", label = 'Largest Growth - Hackney')
min_graph.plot(kind="line", label = 'Smallest Growth - Hounslow')
plt.title('Mean Housing Prices by Year')
plt.xlabel('Year')
plt.ylabel('Mean Price')
plt.legend(loc="upper left")
plt.show()
```
